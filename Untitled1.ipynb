#### Ciencia de Datos

##### La Ciencia de Datos es la disciplina que combina estadística, programación y análisis para transformar grandes volúmenes de datos en información útil, patrones y predicciones que ayudan a tomar mejores decisiones.

#### Mineria de Datos

##### La Minería de Datos (Data Mining) es el proceso técnico de explorar grandes bases de datos de forma automática para descubrir patrones ocultos, correlaciones desconocidas y tendencias que no son visibles a simple vista.
