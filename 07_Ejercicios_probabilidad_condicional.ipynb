# Ejercicio 1: Selección de Representante del Pueblo

## Planteamiento

La población de adultos en un pequeño pueblo que han satisfecho los requisitos para graduarse en la escuela. Se deben clasificar de acuerdo con su sexo y a si trabajan o no actualmente.

### Tabla de Contingencia

|            | Empleado | Desempleado | Total |
|------------|----------|-------------|-------|
| Hombre     | 460      | 40          | 500   |
| Mujer      | 140      | 260         | 400   |
| **Total**  | **600**  | **300**     | **900** |

### Eventos

- **H**: Se escoge un hombre
- **E**: El elegido tiene empleo

### Objetivo

Calcular **P(H|E)**: La probabilidad de que se escoja un hombre dado que el elegido tiene empleo.

```python
# Datos del Ejercicio 1
data_ej1 = {
    'Empleado': [460, 140, 600],
    'Desempleado': [40, 260, 300],
    'Total': [500, 400, 900]
}

df_ej1 = pd.DataFrame(data_ej1, index=['Hombre', 'Mujer', 'Total'])
print("Tabla de Contingencia - Ejercicio 1:")
print(df_ej1)
```

## Solución Ejercicio 1

$$P(H|E) = \frac{P(H \cap E)}{P(E)} = \frac{460}{600}$$

```python
# Cálculo P(H|E)
hombres_empleados = 460
total_empleados = 600

P_H_dado_E = hombres_empleados / total_empleados

print("="*60)
print("RESULTADO EJERCICIO 1")
print("="*60)
print(f"P(H|E) = {hombres_empleados}/{total_empleados} = {P_H_dado_E:.6f}")
print(f"En porcentaje: {P_H_dado_E * 100:.2f}%")
print(f"En fracción: 23/30")
print("="*60)
print("\nInterpretación: Si la persona está empleada, hay un 76.67%")
print("de probabilidad de que sea hombre.")
```

# Ejercicio 2: Probabilidad de Vuelos

## Planteamiento

La probabilidad de que un vuelo de programación regular despegue a tiempo es de **P(D) = 0.83**; la de que llegue a tiempo es de **P(A) = 0.82**; y la de que despegue y llegue a tiempo **P(D ∩ A) = 0.78**.

### Eventos

- **D**: El vuelo despega a tiempo
- **A**: El vuelo llega (arriba) a tiempo

### Objetivos

**a)** Llegue a tiempo dado que despegó a tiempo: **P(A|D)**

**b)** Despegue a tiempo dado que llegó a tiempo: **P(D|A)**

```python
# Datos del Ejercicio 2
P_D = 0.83
P_A = 0.82
P_D_and_A = 0.78

print("Datos del Ejercicio 2:")
print("="*60)
print(f"P(D) = {P_D} (Despega a tiempo)")
print(f"P(A) = {P_A} (Llega a tiempo)")
print(f"P(D ∩ A) = {P_D_and_A} (Despega Y llega a tiempo)")
print("="*60)
```

## Solución Ejercicio 2

### a) P(A|D)

$$P(A|D) = \frac{P(D \cap A)}{P(D)} = \frac{0.78}{0.83}$$

### b) P(D|A)

$$P(D|A) = \frac{P(D \cap A)}{P(A)} = \frac{0.78}{0.82}$$

```python
# Cálculos
P_A_dado_D = P_D_and_A / P_D
P_D_dado_A = P_D_and_A / P_A

print("="*60)
print("RESULTADOS EJERCICIO 2")
print("="*60)
print(f"a) P(A|D) = {P_D_and_A} / {P_D} = {P_A_dado_D:.6f}")
print(f"   En porcentaje: {P_A_dado_D * 100:.2f}%")
print(f"   Interpretación: Si despega a tiempo, hay {P_A_dado_D*100:.2f}%")
print(f"   de probabilidad de que llegue a tiempo.\n")

print(f"b) P(D|A) = {P_D_and_A} / {P_A} = {P_D_dado_A:.6f}")
print(f"   En porcentaje: {P_D_dado_A * 100:.2f}%")
print(f"   Interpretación: Si llega a tiempo, hay {P_D_dado_A*100:.2f}%")
print(f"   de probabilidad de que haya despegado a tiempo.")
print("="*60)
```

```python
# Visualización Ejercicio 2
fig, ax = plt.subplots(figsize=(12, 6))

categorias = ['P(A|D)\nLlega a tiempo\ndado que\ndespegó a tiempo', 
              'P(D|A)\nDespegó a tiempo\ndado que\nllegó a tiempo']
valores = [P_A_dado_D, P_D_dado_A]
colores = ['#e74c3c', '#f39c12']

bars = ax.bar(categorias, valores, color=colores, alpha=0.7, edgecolor='black', width=0.5)
ax.set_ylabel('Probabilidad', fontsize=12)
ax.set_title('Probabilidades Condicionales - Ejercicio 2 (Vuelos)', fontsize=14, fontweight='bold')
ax.set_ylim(0, 1)
ax.grid(axis='y', alpha=0.3)

for bar, val in zip(bars, valores):
    height = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2., height + 0.02,
            f'{val:.4f}\n({val*100:.2f}%)',
            ha='center', va='bottom', fontsize=12, fontweight='bold')

plt.tight_layout()
plt.show()
```

# Ejercicio 3: Hipertensión y Hábito de Fumar

## Planteamiento

En un experimento para estudiar la relación entre la hipertensión y el hábito de fumar, se reunieron los siguientes datos:

### Tabla de Contingencia

|                | No fumadores | Fumadores moderados | Fumadores empedernidos | Total |
|----------------|--------------|---------------------|------------------------|-------|
| Hipertenso     | 21           | 36                  | 30                     | 87    |
| No hipertenso  | 48           | 26                  | 19                     | 93    |
| **Total**      | **69**       | **62**              | **49**                 | **180** |

### Objetivos

Si se selecciona aleatoriamente a uno de estos individuos, encuentre la probabilidad de que la persona:

**a)** Experimente hipertensión, dado que es un fumador empedernido.

**b)** Sea un no fumador, dado que no ha presentado problemas de hipertensión.

```python
# Datos del Ejercicio 3
data_ej3 = {
    'No fumadores': [21, 48, 69],
    'Fumadores moderados': [36, 26, 62],
    'Fumadores empedernidos': [30, 19, 49],
    'Total': [87, 93, 180]
}

df_ej3 = pd.DataFrame(data_ej3, index=['Hipertenso', 'No hipertenso', 'Total'])
print("Tabla de Contingencia - Ejercicio 3:")
print(df_ej3)
```

## Solución Ejercicio 3

### a) P(Hipertenso | Fumador empedernido)

$$P(H|FE) = \frac{\text{Hipertensos fumadores empedernidos}}{\text{Total fumadores empedernidos}} = \frac{30}{49}$$

### b) P(No fumador | No hipertenso)

$$P(NF|NH) = \frac{\text{No fumadores no hipertensos}}{\text{Total no hipertensos}} = \frac{48}{93}$$

```python
# Valores del problema
hipertensos_empedernidos = 30
total_empedernidos = 49

no_fumadores_no_hipertensos = 48
total_no_hipertensos = 93

# Cálculos
P_H_dado_FE = hipertensos_empedernidos / total_empedernidos
P_NF_dado_NH = no_fumadores_no_hipertensos / total_no_hipertensos

print("="*60)
print("RESULTADOS EJERCICIO 3")
print("="*60)
print(f"a) P(Hipertenso|Fumador empedernido)")
print(f"   P(H|FE) = {hipertensos_empedernidos}/{total_empedernidos} = {P_H_dado_FE:.6f}")
print(f"   En porcentaje: {P_H_dado_FE * 100:.2f}%")
print(f"   Interpretación: Si la persona es fumadora empedernida,")
print(f"   hay {P_H_dado_FE*100:.2f}% de probabilidad de que sea hipertensa.\n")

print(f"b) P(No fumador|No hipertenso)")
print(f"   P(NF|NH) = {no_fumadores_no_hipertensos}/{total_no_hipertensos} = {P_NF_dado_NH:.6f}")
print(f"   En porcentaje: {P_NF_dado_NH * 100:.2f}%")
print(f"   Interpretación: Si la persona no es hipertensa,")
print(f"   hay {P_NF_dado_NH*100:.2f}% de probabilidad de que no fume.")
print("="*60)
```

```python
# Visualización 1: Tabla de contingencia como heatmap
fig, ax = plt.subplots(figsize=(12, 6))

heatmap_data = np.array([[21, 36, 30], [48, 26, 19]])

im = ax.imshow(heatmap_data, cmap='YlOrRd', aspect='auto')

ax.set_xticks([0, 1, 2])
ax.set_yticks([0, 1])
ax.set_xticklabels(['No fumadores', 'Fumadores\nmoderados', 'Fumadores\nempedernidos'])
ax.set_yticklabels(['Hipertenso', 'No hipertenso'])

for i in range(2):
    for j in range(3):
        text = ax.text(j, i, str(heatmap_data[i, j]),
                      ha="center", va="center", color="black", fontsize=16, fontweight='bold')

cbar = plt.colorbar(im, ax=ax)
cbar.set_label('Número de Personas', rotation=270, labelpad=20)

ax.set_title('Tabla de Contingencia: Hipertensión vs Hábito de Fumar', 
             fontsize=14, fontweight='bold', pad=20)
ax.set_xlabel('Hábito de Fumar', fontsize=12)
ax.set_ylabel('Condición de Hipertensión', fontsize=12)

plt.tight_layout()
plt.show()
```

```python
# Visualización 2: Probabilidades condicionales
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Gráfico a) P(H|FE)
categorias_a = ['Hipertenso', 'No hipertenso']
valores_a = [hipertensos_empedernidos, total_empedernidos - hipertensos_empedernidos]
colores_a = ['#e74c3c', '#2ecc71']

axes[0].pie(valores_a, labels=categorias_a, autopct='%1.1f%%', colors=colores_a, 
            startangle=90, textprops={'fontsize': 11, 'fontweight': 'bold'})
axes[0].set_title(f'a) Fumadores Empedernidos (n={total_empedernidos})\nP(Hipertenso|Fumador empedernido) = {P_H_dado_FE:.4f}', 
                  fontsize=12, fontweight='bold')

# Gráfico b) P(NF|NH)
categorias_b = ['No fumadores', 'Fumadores\nmoderados', 'Fumadores\nempedernidos']
valores_b = [48, 26, 19]
colores_b = ['#3498db', '#f39c12', '#e74c3c']

bars = axes[1].bar(categorias_b, valores_b, color=colores_b, alpha=0.7, edgecolor='black')
axes[1].set_ylabel('Número de Personas', fontsize=11)
axes[1].set_title(f'b) Personas No Hipertensas (n={total_no_hipertensos})\nP(No fumador|No hipertenso) = {P_NF_dado_NH:.4f}', 
                  fontsize=12, fontweight='bold')
axes[1].grid(axis='y', alpha=0.3)

for bar, val in zip(bars, valores_b):
    height = bar.get_height()
    axes[1].text(bar.get_x() + bar.get_width()/2., height + 1,
                str(val), ha='center', va='bottom', fontsize=11, fontweight='bold')

plt.tight_layout()
plt.show()
```

## Análisis Adicional - Ejercicio 3

```python
# Calcular tasas de hipertensión por grupo
print("Tasas de Hipertensión por Grupo de Fumadores:")
print("="*60)

P_H_dado_NF = 21 / 69  # No fumadores
P_H_dado_FM = 36 / 62  # Fumadores moderados
P_H_dado_FE_calc = 30 / 49  # Fumadores empedernidos

print(f"P(Hipertenso|No fumador) = {P_H_dado_NF:.4f} ({P_H_dado_NF*100:.2f}%)")
print(f"P(Hipertenso|Fumador moderado) = {P_H_dado_FM:.4f} ({P_H_dado_FM*100:.2f}%)")
print(f"P(Hipertenso|Fumador empedernido) = {P_H_dado_FE_calc:.4f} ({P_H_dado_FE_calc*100:.2f}%)")

print("\nObservación: La tasa de hipertensión aumenta con el nivel de tabaquismo.")
print(f"Incremento de no fumador a empedernido: {(P_H_dado_FE_calc - P_H_dado_NF)*100:.2f} puntos porcentuales")
print("="*60)
```

```python
# Visualización comparativa de tasas de hipertensión
fig, ax = plt.subplots(figsize=(10, 6))

grupos = ['No\nfumadores', 'Fumadores\nmoderados', 'Fumadores\nempedernidos']
tasas = [P_H_dado_NF, P_H_dado_FM, P_H_dado_FE_calc]
colores_grad = ['#2ecc71', '#f39c12', '#e74c3c']

bars = ax.bar(grupos, tasas, color=colores_grad, alpha=0.7, edgecolor='black')
ax.set_ylabel('Probabilidad de Hipertensión', fontsize=12)
ax.set_title('Tasa de Hipertensión por Nivel de Tabaquismo', fontsize=14, fontweight='bold')
ax.set_ylim(0, 0.8)
ax.grid(axis='y', alpha=0.3)

for bar, val in zip(bars, tasas):
    height = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2., height + 0.02,
            f'{val:.4f}\n({val*100:.2f}%)',
            ha='center', va='bottom', fontsize=11, fontweight='bold')

plt.tight_layout()
plt.show()
```

## Resumen Ejercicio 3

### Respuestas Finales

**a) P(Hipertenso|Fumador empedernido) = 30/49 ≈ 0.6122 (61.22%)**

Si la persona es fumadora empedernida, hay una probabilidad del **61.22%** de que sea hipertensa.

**b) P(No fumador|No hipertenso) = 48/93 ≈ 0.5161 (51.61%)**

Si la persona no es hipertensa, hay una probabilidad del **51.61%** de que no fume.

### Conclusiones

1. **Relación entre tabaquismo e hipertensión**: Los datos muestran una clara relación positiva entre el nivel de tabaquismo y la hipertensión:
   - No fumadores: 30.43% de hipertensión
   - Fumadores moderados: 58.06% de hipertensión
   - Fumadores empedernidos: 61.22% de hipertensión

2. **Incremento del riesgo**: Los fumadores empedernidos tienen aproximadamente el doble de probabilidad de ser hipertensos comparado con los no fumadores.

3. **Población no hipertensa**: Entre las personas sin hipertensión, aproximadamente la mitad son no fumadores, lo que sugiere que no fumar está asociado con menor riesgo de hipertensión.
