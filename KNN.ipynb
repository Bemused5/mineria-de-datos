## 🟢 Ejercicio 1: k-NN manual (k = 1, distancia euclidiana)

**Objetivo didáctico:** Entender cómo funciona k-NN desde cero.

En este ejercicio, clasificaremos un punto $P$ basándonos en el vecino más cercano utilizando la distancia euclidiana.

```python
import math
 
# Datos de entrenamiento
X = [(2,4), (4,4), (4,2), (6,2)]
y = ['Rojo', 'Rojo', 'Azul', 'Azul']
 
# Punto a clasificar
P = (3, 3)

def distancia_euclidiana(p1, p2):
    """Calcula la distancia en línea recta entre dos puntos."""
    return math.sqrt((p1[0]-p2[0])**2 + (p1[1]-p2[1])**2)

# Cálculo del vecino más cercano
distancias = []
for i in range(len(X)):
    d = distancia_euclidiana(P, X[i])
    distancias.append((d, y[i]))
 
# Ordenar por distancia (el más cercano primero)
distancias.sort(key=lambda x: x[0])
 
print("Lista de distancias:", distancias)
print("Clase asignada (k=1):", distancias[0][1])
```

## 🟡 Ejercicio 2: k-NN manual con k = 3

**Tema clave:** Votación por mayoría.

Cuando $k > 1$, la clase se decide por la mayoría entre los $k$ vecinos más cercanos.

```python
from collections import Counter
 
k = 3
vecinos = distancias[:k]
 
# Extraemos solo las etiquetas de los vecinos
clases = [clase for _, clase in vecinos]
# Contamos la frecuencia de cada clase y tomamos la más común
resultado = Counter(clases).most_common(1)
 
print("Vecinos cercanos:", vecinos)
print("Clase asignada por mayoría:", resultado[0][0])
```

## 🔵 Ejercicio 3: Distancia Manhattan

**Aprendizaje:** Diferentes métricas de distancia.

La distancia Manhattan (o City Block) suma las diferencias absolutas de las coordenadas. Es útil en ciertos tipos de datos donde los ejes no son comparables directamente.

```python
def distancia_manhattan(p1, p2):
    return abs(p1[0]-p2[0]) + abs(p1[1]-p2[1])
 
X = [(1,1), (2,3), (5,4), (6,5)]
y = ['Baja', 'Baja', 'Alta', 'Alta']
P = (4,3)
 
distancias_m = []
 
for i in range(len(X)):
    d = distancia_manhattan(P, X[i])
    distancias_m.append((d, y[i]))
 
distancias_m.sort(key=lambda x: x[0])
 
print("Distancias Manhattan:", distancias_m)
print("Clase asignada:", distancias_m[0][1])
```

## 🟣 Ejercicio 4: k-NN con scikit-learn

**Conexión profesional:** Uso de librerías estándar de la industria.

Utilizaremos un subconjunto simplificado del dataset Iris para clasificar una nueva muestra.

```python
import numpy as np
from sklearn.neighbors import KNeighborsClassifier

# Datos tipo Iris (reducido)
X_iris = np.array([
    [5.1, 3.5],
    [4.9, 3.0],
    [6.2, 3.4],
    [6.0, 2.9]
])
 
y_iris = np.array(['Setosa', 'Setosa', 'Versicolor', 'Versicolor'])
 
P_nuevo = np.array([[5.8, 3.2]])

# Entrenamiento y predicción
# n_neighbors es el parámetro 'k'
modelo = KNeighborsClassifier(n_neighbors=3)
modelo.fit(X_iris, y_iris)
 
prediccion = modelo.predict(P_nuevo)
print("Clase predicha por scikit-learn:", prediccion[0])
```

## ✅ Ejercicio 5: Impacto de la normalización en k-NN

**Objetivo:** Mostrar cómo la escala de las variables afecta la clasificación y por qué es imprescindible normalizar.

### Dataset: Clientes para un crédito
| Cliente | Edad (años) | Ingreso mensual ($) | Riesgo |
|---------|-------------|---------------------|--------|
| 1       | 25          | 8,000               | Alto   |
| 2       | 30          | 12,000              | Alto   |
| 3       | 45          | 45,000              | Bajo   |
| 4       | 50          | 60,000              | Bajo   |
| 5       | 35          | 15,000              | Alto   |
| 6       | 48          | 55,000              | Bajo   |

**Punto a clasificar:** Nuevo cliente (Edad = 40, Ingreso = 20,000)

### 🟥 Parte A: k-NN SIN normalización

```python
# Dataset
X_clientes = np.array([
    [25, 8000],
    [30, 12000],
    [45, 45000],
    [50, 60000],
    [35, 15000],
    [48, 55000]
])
 
y_riesgo = np.array(['Alto', 'Alto', 'Bajo', 'Bajo', 'Alto', 'Bajo'])
 
nuevo_cliente = np.array([[40, 20000]])
 
# Modelo k-NN
modelo_sin = KNeighborsClassifier(n_neighbors=3)
modelo_sin.fit(X_clientes, y_riesgo)
 
# Predicción
res_sin = modelo_sin.predict(nuevo_cliente)
print("Predicción SIN normalizar:", res_sin[0])
```

**¿Qué está pasando aquí?**
- La distancia está dominada casi totalmente por el **ingreso**.
- La edad (25-50) tiene muy poco impacto comparada con el ingreso (8,000-60,000).
- El modelo busca clientes con ingresos cercanos a 20,000, ignorando parcialmente la edad.

### 🟢 Parte B: k-NN CON normalización
Usaremos `StandardScaler` para que ambas variables tengan media 0 y desviación estándar 1.

```python
from sklearn.preprocessing import StandardScaler
 
# Escalamiento
scaler = StandardScaler()
X_norm = scaler.fit_transform(X_clientes)
nuevo_norm = scaler.transform(nuevo_cliente)
 
# Modelo k-NN con datos normalizados
modelo_norm = KNeighborsClassifier(n_neighbors=3)
modelo_norm.fit(X_norm, y_riesgo)
 
# Predicción
res_norm = modelo_norm.predict(nuevo_norm)
print("Predicción CON normalización:", res_norm[0])
```

### 🧠 Comparación conceptual

| Modelo | Usa edad correctamente | Usa ingreso correctamente | Confiable |
|--------|------------------------|---------------------------|-----------|
| Sin normalizar | ❌ | ✅ | ❌ |
| Normalizado | ✅ | ✅ | ✅ |

## ⚠️ Ejercicio 6: Empates

**Discusión:** ¿Qué sucede cuando hay la misma cantidad de vecinos de diferentes clases?

```python
X_empate = [(2,2), (2,4), (4,2), (4,4)]
y_empate = ['A', 'B', 'B', 'A']
P_empate = (3,3)
 
distancias_e = [(distancia_euclidiana(P_empate, X_empate[i]), y_empate[i]) for i in range(len(X_empate))]
distancias_e.sort(key=lambda x: x[0])
 
k_empate = 4
clases_vecinas = [clase for _, clase in distancias_e[:k_empate]]
 
print("Clases de los 4 vecinos más cercanos:", clases_vecinas)
print("¡Hay un empate 2 vs 2!")
```

**Soluciones ante un empate:**
1. Reducir $k$ (por ejemplo, usar $k$ impar).
2. Usar ponderación por distancia (los más cercanos valen más).
3. Elegir la clase prioritariamente según el orden del dataset o la frecuencia global.
